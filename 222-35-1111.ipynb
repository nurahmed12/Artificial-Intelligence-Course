```python
class Fruit:
    def __init__(self, name, color):
        self.name = name
        self.color = color

    def introduce(self):
        print(f"I am a {self.name} and my color is {self.color}.")


class Apple(Fruit):
    def __init__(self, variety):
        # Call the parent class's __init__
        super().__init__("Apple", "Red")  # Apples are generally red
        self.variety = variety

    def introduce(self):
        # Call the parent class's introduce
        super().introduce()
        print(f"I am a {self.variety} apple.")

fruit1 = Fruit("Banana", "Yellow")
fruit1.introduce()

apple1 = Apple("Fuji")
apple1.introduce()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
heights = [5, 10, 15, 20]
x_pos = np.arange(len(heights))  # X positions for the bars

# Create the bar chart
plt.barh(x_pos, heights)

# Add labels and title
plt.xticks(x_pos, ['A', 'B', 'C', 'D'])  # Replace with your desired labels
plt.xlabel("Categories")
plt.ylabel("Heights")
plt.title("Bar Chart")

# Show the chart
plt.show()
```

```python
def f(x):
  return x**2 + 5*x + 6

def df(x):
  return 2*x + 5  # Derivative of f(x)

def gradient_descent(learning_rate, num_iterations, initial_x):
  x = initial_x
  for i in range(num_iterations):
    x = x - learning_rate * df(x)
  return x

# Set hyperparameters
learning_rate = 0.1
num_iterations = 1000
initial_x = 0

# Run gradient descent
x_min = gradient_descent(learning_rate, num_iterations, initial_x)

print("Minimum found at x =", x_min)
print("Minimum value of f(x) =", f(x_min))
```
