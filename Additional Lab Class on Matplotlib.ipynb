### Line plot

```python
# Creating an empty list
l1 = []

# Adding elements to the list
for i in range(10):
  l1.append(i/2)
l1
```

```python
# Generating numbers between 1 to 39
x = range(1, 40)
y = [i/2 for i in x]


plt.plot(x, y, 'r')
plt.show()
```

```python
# create a plot figure
fig = plt.figure() # figure is the final object

x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x)) # '-'
```

```python
plt.plot(x, np.cos(x)) # '--' # move it to previous cell
```

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

# can be done in a line
plt.plot(x, np.sin(x), x, np.cos(x))

plt.xlabel('x')
plt.ylabel('y')
plt.title("Sin/Cos Plot")
plt.legend(['sin(x)', 'cos(x)']) # loc=1

plt.grid(True) # adding grid

plt.show()
```

```python
x = np.linspace(0, 10, 10)
# create a plot figure
fig = plt.figure(figsize=(8, 8))

# create 4 subplots (rows, cols, panel)
plt.subplot(2, 2, 1)
plt.plot(x, x, 'o')

plt.subplot(2, 2, 2)
plt.plot(x, x**3)

plt.subplot(2, 2, 3)
plt.plot(x, np.sin(x))

plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x))

#plt.title('Mixed Plot')
```
