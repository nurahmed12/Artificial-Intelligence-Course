Maximize the cost function for f(x) = -x^2 + 4x

```python
def f(x):
  return (-x**2) + (4 * x)
```

```python
# Initialize x = x0 (initial solution)
x = 3
```

```python
# Evaluate f(x)
f(x)
```

```python
import math
step_size = 0.1

# Repeat:
for i in range(100):
  x_curr = f(x)

  # Generate neighbors of x
  left_x = x - step_size
  right_x = x + step_size

  # Evaluate f(neighbor) for each neighbor
  eval_left_x = f(left_x)
  eval_right_x = f(right_x)

  # Find the neighbor with the best f(neighbor)
  # If f(neighbor) is better than f(x)
    # Move to the neighbor x = neighbor
  if eval_left_x > x_curr and eval_left_x > eval_right_x:
      x = left_x
  elif eval_right_x > x_curr and eval_right_x > eval_left_x:
      x = right_x
  # Else:
    # Terminate and return x as the solution
  else:
    break

  print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
```
