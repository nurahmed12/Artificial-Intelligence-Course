**Comparison Operators**


*   Bitwise Operators (|, &):
Operate at the bit level on integers or booleans.

| (Bitwise OR): Combines bits where at least one is 1.
& (Bitwise AND): Combines bits where both are 1.
Logical Operators (or, and):
Operate based on truthiness of values.

or: Returns the first truthy value (or the last if none are truthy).
and: Returns the first falsy value (or the last if all are truthy).
Key Difference:

Bitwise operators process entire binary values.
Logical operators use short-circuit evaluation and work on any object.



```python
'hello' == 'hi'
```

```python
'hi' == 'hi'
```
